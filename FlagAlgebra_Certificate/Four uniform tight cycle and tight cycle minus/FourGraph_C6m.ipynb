```python
import pickle, glob
from fractions import Fraction

hits = glob.glob("**/FourGraph_C6m_6vtx_opt.pickle", recursive=True)
if not hits:
    raise FileNotFoundError("Cannot find FourGraph_C6m_6vtx_opt.pickle")
cert_file = hits[0]
print("using certificate file:", cert_file)

with open(cert_file, "rb") as f:
    cert = pickle.load(f)

base_flags = cert["base flags"]
slack_vector = cert["slack vector"]

assert len(base_flags) == len(slack_vector)

def edges_of_flag_tuple(F):
    blocks = dict(F[2])
    return tuple(blocks.get("edges", ()))

forced_zero = []
sharp = []

for i, (F, c) in enumerate(zip(base_flags, slack_vector)):
    c = Fraction(c)

    rec = {
        "index": i,
        "num_vertices": F[0],
        "num_edges": len(edges_of_flag_tuple(F)),
        "edges": edges_of_flag_tuple(F),
        "slack": c,
    }

    if c.numerator == 0:
        sharp.append(rec)
    elif c.numerator > 0:
        forced_zero.append(rec)
    else:
        raise ValueError(f"Negative slack at index {i}: {c}")

print("certificate result =", cert["result"])
print("target size        =", cert["target size"])
print("number of base flags =", len(base_flags))
print("forced-zero flags    =", len(forced_zero))
print("sharp flags          =", len(sharp))
```

```python
from itertools import combinations, product

def edge_set_of_flag_tuple(F):
    blocks = dict(F[2])
    return set(tuple(e) for e in blocks.get("edges", ()))

def template_edge_from_colors(cols):
    # allowed edge types: 0001 or 0111
    s = sum(cols)
    return s == 1 or s == 3

def valid_template_colorings(F):
    """
    Return all 0/1-colorings witnessing that F appears in the bipartite
    4-graph template with allowed edge-types 0001 and 0111.
    """
    n = F[0]
    E = edge_set_of_flag_tuple(F)
    all4 = list(combinations(range(n), 4))
    sols = []

    for coloring in product([0, 1], repeat=n):
        ok = True
        for S in all4:
            templ_edge = template_edge_from_colors([coloring[i] for i in S])
            inH = tuple(S) in E
            if templ_edge != inH:
                ok = False
                break
        if ok:
            sols.append(coloring)
    return sols

def appears_in_extremal_template(rec):
    F = base_flags[rec["index"]]
    sols = valid_template_colorings(F)
    return len(sols) > 0, sols

sharp_in_template = []
sharp_not_in_template = []

for rec in sharp:
    ok, sols = appears_in_extremal_template(rec)
    if ok:
        sharp_in_template.append((rec, sols))
    else:
        sharp_not_in_template.append(rec)

forced_zero_in_template = []
forced_zero_not_in_template = []

for rec in forced_zero:
    ok, sols = appears_in_extremal_template(rec)
    if ok:
        forced_zero_in_template.append((rec, sols))
    else:
        forced_zero_not_in_template.append(rec)

print("sharp in template        =", len(sharp_in_template))
print("sharp not in template    =", len(sharp_not_in_template))
print("forced_zero in template  =", len(forced_zero_in_template))
print("forced_zero not template =", len(forced_zero_not_in_template))
```

```python
def tuple_to_flag(F, FG):
    n = F[0]
    blocks = dict(F[2])
    edges = [list(e) for e in blocks.get("edges", ())]
    return FG(n, edges=edges)

FG = FourGraphTheory
FG.reset()
FG.exclude([C6m])

constr = FG.blowup_construction(
    6,
    2,
    edges=[[0,0,0,1], [0,1,1,1]]
)

template_densities = []
for rec in sharp + forced_zero:
    F_obj = tuple_to_flag(base_flags[rec["index"]], FG)
    d = constr.density(F_obj)
    template_densities.append((rec["index"], rec["slack"], d, rec["edges"]))

for x in template_densities:
    print(x)
```
