```python
# Setting up {K_4^{3-}, C_5^{3-}}-free 3-graph theory
K4m = ThreeGraphTheory.p(4, edges=[[0, 1, 2], [0, 2, 3], [0, 1, 3]])
C5m = ThreeGraphTheory.p(5, edges=[[0, 1, 2], [1, 2, 3], [2, 3, 4], [3, 4, 0]])
ThreeGraphTheory.exclude([K4m, C5m])
TG = ThreeGraphTheory

print("Number of structures without K4m, C5m")
print("and size 5: ", len(TG.generate(5)))
print("and size 6: ", len(TG.generate(6)))
print("and size 7: ", len(TG.generate(7)))
```

```python
# Proposition 3.1: this part shows that every 3-graph with S_2-density at least 6/13 - 10^{-6} contains a large 3-partite subgraph

target_size_1 = 7
assume_1 = []

f222 = TG.p(6, ftype=[0, 1, 2], edges=[[0, 1, 2], [3, 4, 5], [0, 1, 5], [0, 2, 4], [1, 2, 3]]).afae()
k222 = f222.project()

# Define induced S2
S3 = TG(5, edges=[[0, 1, 2], [0,1,3],[0,1,4]]) + TG(5,edges =[ [ 0, 1, 4 ], [ 0, 1, 3 ], [ 0, 1, 2 ], [ 2, 3, 4 ] ])
# S2-density lower bound assumption
assume_1 += [S3 - 1/4 + 1/1000000]

gamma = TG.optimize(k222, target_size_1, maximize=False, positives=assume_1, exact=True, 
                     file="C5MinusL3Norm_Part1_cut_ratio_rounded", 
                     denom=1024*1024
                    )
# a31 is a lower bound for the max-cut ratio. Here, 1/4 is the Turan density of {K4m, C5m}, i.e. the upper bound for the edge density
a31 = gamma/(1/4)  

# Verifying the computation
TG.verify("C5MinusL2Norm_Part1_cut_ratio_rounded", target_element=k222)

print("The max-cut ratio is at least {} ~= {}".format(a31, a31.n()))
```
